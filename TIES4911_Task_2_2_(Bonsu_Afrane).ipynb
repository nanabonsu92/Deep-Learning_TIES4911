```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.utils import to_categorical
```

```python
# Load the Wine Quality dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
data = pd.read_csv(url, sep=';')
```

```python
data
```

```python
# Convert quality to categorical data
bins = [2, 5.5, 7.5, 8]
group_names = ['low', 'medium', 'high']
data['quality'] = pd.cut(data['quality'], bins=bins, labels=group_names)
```

```python
# Label encode the categorical data
label_quality = LabelEncoder()
data['quality'] = label_quality.fit_transform(data['quality'])
```

```python
# Split data into features and target
X = data.drop('quality', axis=1)
y = data['quality']
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Convert labels to categorical
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
```

```python
# Define the model
model = Sequential([
    Dense(256, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')  # Assuming 3 classes
])
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=50, validation_split=0.2)
```

```python
# Evaluate the model and print the results
evaluation_results = model.evaluate(X_test, y_test)
print(f"Test Loss: {evaluation_results[0]}")
print(f"Test Accuracy: {evaluation_results[1]}")
```

```python
# Make predictions
predictions = model.predict(X_test)
predicted_classes = np.argmax(predictions, axis=1)
actual_classes = np.argmax(y_test, axis=1)

# Display some predictions
for i in range(10):
    print(f"Actual Quality: {actual_classes[i]}, Predicted Quality: {predicted_classes[i]}")
```
