```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the dataset
dataframe = pd.read_csv("/content/gdrive/MyDrive/Deep Learning 2024/Task 1/Automobile_price_data_Raw_set.csv")
```

```python
dataframe
```

```python
# Dropping rows with any missing values (NaNs)
df = dataframe.dropna()
```

```python
df.describe()
```

```python
feature_columns = ['make','body-style', 'wheel-base', 'engine-size', 'horsepower', 'peak-rpm', 'highway-mpg' ]  # Using 'make','body-style', 'wheel-base', 'engine-size', 'horsepower', 'peak-rpm', 'highway-mpg'
target_column = 'price'    # Target: 'price'
```

```python
# Extracting features and target
X = df[feature_columns]
Y = df[target_column]
```

```python
# One-hot encoding the categorical data
X_encoded = pd.get_dummies(X, columns=['make', 'body-style'])
```

```python
## Splitting the data
x_train, x_test, y_train, y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=0)
```

```python
# Initializing MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Scaling features
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Scaling target
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))
```

```python
#Building the DNN model
class DeepNeuralNetworkModel:
    def __init__(self):
        self.model = tf.keras.Sequential([
            # Input layer with the number of features from your data
            tf.keras.layers.Dense(128, activation='relu', input_shape=(x_train_scaled.shape[1],)),
            # Additional hidden layers
            tf.keras.layers.Dense(64, activation='relu'),
            tf.keras.layers.Dense(32, activation='relu'),
            # Output layer with a single neuron as you are performing regression
            tf.keras.layers.Dense(1)
        ])
        self.model.compile(optimizer='adam', loss='mse')

    def train(self, X, y, epochs=100, batch_size=32):
        self.model.fit(X, y, epochs=epochs, batch_size=batch_size)

    def predict(self, X):
        return self.model.predict(X)
```

```python
# Creating and training the DNN model
dnn_model = DeepNeuralNetworkModel()
dnn_model.train(x_train_scaled, y_train_scaled, epochs=100)
```

```python
# Training the Scikit-learn model
sk_linear_model = LinearRegression()
sk_linear_model.fit(x_train_scaled, y_train_scaled)
```

```python
# Defining the input data
input_data = pd.DataFrame([['audi', 'hatchback', 99.5, 131, 160, 5500, 22]],
                          columns=['make', 'body-style', 'wheel-base', 'engine-size', 'horsepower', 'peak-rpm', 'highway-mpg'])
```

```python
# One-hot encoding and aligning the input data
input_data_encoded = pd.get_dummies(input_data, columns=['make', 'body-style'])
input_data_encoded = input_data_encoded.reindex(columns=x_train.columns, fill_value=0)
```

```python
# Scaling the input data
input_scaled = scaler.transform(input_data_encoded)
```

```python
# Converting input_scaled to float32 if it's not already
input_scaled_2 = np.array(input_scaled, dtype=np.float32)
```

```python
# Predicting with DNN model
predicted_price_dnn_scaled = dnn_model.predict(input_scaled_2)
predicted_price_dnn = y_scaler.inverse_transform(predicted_price_dnn_scaled)
```

```python
# Predicting with Scikit-learn model
predicted_price_sk_scaled = sk_linear_model.predict(input_scaled_2)
predicted_price_sk = y_scaler.inverse_transform(predicted_price_sk_scaled)
```

```python
print("Predicted price (Deep Neural Network): ", predicted_price_dnn[0])
print("Predicted price (Scikit-learn): ", predicted_price_sk[0])
```

```python
# Converting test data to float32
x_test_scaled = np.array(x_test_scaled, dtype=np.float32)
```

```python
# Calculating MSE for both models
mse_dnn = mean_squared_error(y_test_scaled, dnn_model.predict(x_test_scaled))
mse_sk = mean_squared_error(y_test_scaled, sk_linear_model.predict(x_test_scaled))

print("MSE (Deep Neural Network): ", mse_dnn)
print("MSE (Scikit-learn): ", mse_sk)
```
